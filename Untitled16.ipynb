```python
def assign_members(members):
    queue_even = []
    queue_odd = []

    for member in members:
        if member % 2 == 0:
            queue_even.append(member)
        else:
            queue_odd.append(member)

    print("Volunteer A:", end=" ")
    for member in queue_even:
        print(member, end=" ")
    print()

    print("Volunteer B:", end=" ")
    for member in queue_odd:
        print(member, end=" ")
    print()
assign_members([2, 4, 7, 10, 1, 9, 11])
```

```python
class Queue():
    def __init__(self):
        self.data = []

    def isempty(self):
        return not len(self.data)

    def enqueue(self, element):
        self.data.append(element)

    def dequeue(self):
        if self.isempty():
            return "Warning: The queue is empty"
        self.data.pop(0)

    def peek(self):
        if self.isempty():
            return "Warning: The queue is empty"
        return self.data[0]
    
    def display(self):
       if self.isempty():
        print("Queue is empty")
       else:
        print(f"{self.data}")  
p=Queue()
A=Queue()
B=Queue()
k=[2 ,4 ,7 ,10 ,1 ,9 ,11]
for i in k:
  p.enqueue(i)

for m in p.data:
  if m%2==0:
    A.enqueue(m)
  else:
    B.enqueue(m)
print("A : ",end="")
A.display()
print("B : ",end="")
B.display()
```
