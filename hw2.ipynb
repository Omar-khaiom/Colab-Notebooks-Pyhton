```python
class Stack:
    def __init__(self):
        self.__data = []
    def size(self):
        return len(self.__data)
    def isempty(self):
        if self.size() == 0:
            return True
        else:
            return False
    def push(self, data):
        self.__data.append(data)
    def pop(self):
        if self.isempty():
            print("Stack is empty")
        else:
            self.__data = self.__data[:-1]
    def peek(self):
        if self.isempty():
            print("Stack is empty")
        else:
            return self.__data[-1]
    def printStack(self):
        if self.isempty():
            print("Stack is empty")
        else:
            print(self.__data)
stack_1 = Stack()
stack_1.push(1)
stack_1.push(5)
stack_1.push(11)
stack_1.push(3)
stack_1.printStack()

stack_2 = Stack()
stack_2.push(4)
stack_2.push(18)
stack_2.push(1)
stack_2.push(-2)
stack_2.printStack()

def diff():
    sum_1 = 0
    sum_2 = 0
    if stack_1.isempty() or stack_2.isempty():
        print("One of the stacks is empty!!!")
        return
    while not stack_1.isempty():
        temp = stack_1.peek()
        sum_1 += temp
        stack_1.pop()
    while not stack_2.isempty():
        temp = stack_2.peek()
        sum_2 += temp
        stack_2.pop()
    differ = sum_1 - sum_2
    if differ == 0:
        print("Stacks are already balanced")
        stack_1.printStack()
        stack_2.printStack()
    elif differ > 0:
        stack_2.push(differ)
        print("Adding", differ, "to stack B")
        stack_1.printStack()
        stack_2.printStack()
    else:
        stack_1.push(-differ)
        print("Adding", -differ, "to stack A")
        stack_1.printStack()
        stack_2.printStack()
diff() 
```
