```python
#4. Write findMinR(self, root) that will return the minimum of a BST using recursion.

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
def insert(node, data):
    if node is None :
        return Node(data)
    else:
        if data <= node.data:
            node.left = insert(node.left, data)
        else:
            node.right = insert(node.right, data)

        return node
def minValue(node):
    if node.left == None:
        return node.data
    return minValue(node.left)

if __name__ == "__main__" :
    root = None
    root = insert(root, 42)
    insert(root, 21)
    insert(root, 31)
    insert(root, 32)
    insert(root, 61)
    insert(root, 51)

    print(minValue(root))
```

```python
#6. Check whether the leaf nodes of a BST is even or odd.

#10

#/ \

#5 13

#/ \ \

#3 7 14

#The output of the BST above is [odd, odd, even].

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
def isLeaf(node):
    return node is not None and node.left is None and node.right is None
def checkLeafNodes(root):
    if not root:
        return []
    result = []
    def dfs(node):
        if isLeaf(node):
            result.append("even" if node.val % 2 == 0 else "odd")
            return
        if node.left:
            dfs(node.left)
        if node.right:
            dfs(node.right)
    dfs(root)
    return result
def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)

    leaf_nodes_parity = checkLeafNodes(root)
    print("The leaf nodes are:", leaf_nodes_parity)

if __name__ == "__main__":
    main()
```

```python
#10. Write a function int countLeafNodes(root) to calculate number of all leaf nodes
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def countLeafNodes(root):
    if not root:
        return 0
    if not root.left and not root.right:
        return 1
    return countLeafNodes(root.left) + countLeafNodes(root.right)

def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)
    num_leaf_nodes = countLeafNodes(root)
    print("Number of leaf nodes in the BST:", num_leaf_nodes)
if __name__ == "__main__":
    main()
```
