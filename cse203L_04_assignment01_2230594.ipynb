```python


class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next
#1. Given two sorted linked lists, write a function to merge them into a single sorted list.
def merge_sorted_lists(list1, list2):
    temp = ListNode()
    current = temp

    while list1 is not None and list2 is not None:
        if list1.value < list2.value:
            current.next = ListNode(list1.value)
            list1 = list1.next
        else:
            current.next = ListNode(list2.value)
            list2 = list2.next

        current = current.next

    while list1 is not None:
        current.next = ListNode(list1.value)
        list1 = list1.next
        current = current.next

    while list2 is not None:
        current.next = ListNode(list2.value)
        list2 = list2.next
        current = current.next

    return temp.next


list1 = ListNode(1, ListNode(2, ListNode(4)))
list2 = ListNode(1, ListNode(3, ListNode(4)))

merged_list = merge_sorted_lists(list1, list2)


while merged_list is not None:
    print(merged_list.value, end=" -> ")
    merged_list = merged_list.next

```

```python
#2. Write a function to check if a singly linked list is a palindrome or not. The function will return True or False

def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] == s[-1]:
        return is_palindrome(s[1:-1])
    return False

input_list = [1,2,3,4,5]
result = is_palindrome(input_list)
print(f"Is '{input_list}' a palindrome? {result}")
```

```python

#3. Write a function to Find the intersection point of two linked lists.
def find_intersection_point(headA, headB):
    def get_length(head):
        length = 0
        while head:
            length += 1
            head = head.next
        return length

    lenA, lenB = get_length(headA), get_length(headB)

    while lenA > lenB:
        headA = headA.next
        lenA -= 1

    while lenB > lenA:
        headB = headB.next
        lenB -= 1

    while headA is not None and headB is not None:
        if headA == headB:
            return headA

        headA = headA.next
        headB = headB.next

    return None

common_node = ListNode(8, ListNode(4, ListNode(5)))
listA = ListNode(4, ListNode(1, common_node))
listB = ListNode(5, ListNode(0, ListNode(1, common_node)))

intersection_point = find_intersection_point(listA, listB)

if intersection_point:
    print(f"Intersection Point : {intersection_point.value}")
else:
    print("There is No Intersection Point")



```

```python
#4. Write a function to remove all the nodes with a given value for a singly linked
def remove_elements(head, val):
    dummy = ListNode(0)
    dummy.next = head
    current = dummy
    while current.next is not None:
        if current.next.value == val:
            current.next = current.next.next
        else:
            current = current.next

    return dummy.next
original_list = ListNode(1, ListNode(2, ListNode(6, ListNode(3, ListNode(4, ListNode(5, ListNode(6)))))))

new_list = remove_elements(original_list, 6)
while new_list is not None:
    print(new_list.value, end=" -> ")
    new_list = new_list.next
```

```python
#5. Reverse a given singly linked list.
def reverse_linked_list(head):
    prev = None
    current = head
    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev

original_list = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
reversed_list = reverse_linked_list(original_list)
while reversed_list is not None:
    print(reversed_list.value, end=" -> ")
    reversed_list = reversed_list.next

```

```python
#1. Write a function that rotates a doubly linked list by a given number of positions to the right Example:


class DoublyListNode:
    def __init__(self, value=0, prev=None, next=None):
        self.value = value
        self.prev = prev
        self.next = next

def rotate_doubly_linked_list(dll_head, k):
    if not dll_head or k == 0:
        return dll_head

    length = 0
    current = dll_head
    while current.next:
        length += 1
        current = current.next

    k = k % length

    if k == 0:
        return dll_head
    current = dll_head
    for _ in range(length - k - 1):
        current = current.next
    new_head = current.next
    current.next = None
    new_head.prev = None

    current = new_head
    while current.next:
        current = current.next

    current.next = dll_head
    dll_head.prev = current

    return new_head

dll_head = DoublyListNode(1)
dll_head.next = DoublyListNode(2, dll_head)
dll_head.next.next = DoublyListNode(3, dll_head.next)
dll_head.next.next.next = DoublyListNode(4, dll_head.next.next)
dll_head.next.next.next.next = DoublyListNode(5, dll_head.next.next.next)
dll_head.next.next.next.next.next = DoublyListNode(6, dll_head.next.next.next.next)

rotated_dll_head = rotate_doubly_linked_list(dll_head, 2)

current = rotated_dll_head
while current:
    print(current.value, end=" <-> ")
    current = current.next
```

```python
#2. Find the Nth node from the end of the doubly linked list.
def nth_node_from_end(dll_head, n):
    if not dll_head or n <= 0:
        return None

    first = dll_head
    second = dll_head
    for _ in range(n):
        if second is None:
            return None
        second = second.next

    while second.next:
        first = first.next
        second = second.next

    return first
dll_head = DoublyListNode(1)
dll_head.next = DoublyListNode(2, dll_head)
dll_head.next.next = DoublyListNode(3, dll_head.next)
dll_head.next.next.next = DoublyListNode(4, dll_head.next.next)
n = 2
nth_node = nth_node_from_end(dll_head, n)
if nth_node:
    print(f"Node with value {nth_node.value}")
else:
    print(f"There is no {n}th node from the end.")
```

```python
#1. Write a function to check if a stack is palindrome or not. The function will return True or False
def is_stack_palindrome(stack):
    temp_stack = []
    for item in stack:
        temp_stack.append(item)
    while stack:
        if stack.pop() != temp_stack.pop():
            return False
    return True

stack1 = [1, 2, 3, 4, 5]
print(is_stack_palindrome(stack1))
stack2 = [1, 2, 3, 2, 1]
print(is_stack_palindrome(stack2))
```

```python

#2. Write a function to remove the maximum value from the stack.
def remove_max_from_stack(stack):
    if not stack:
        return []
    max_value = max(stack)
    temp_stack = []
    while stack:
        if stack[-1] == max_value:
            stack.pop()
            break
        temp_stack.append(stack.pop())
    while temp_stack:
        stack.append(temp_stack.pop())

    return stack
stack = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
result_stack = remove_max_from_stack(stack)
print(result_stack)
```

```python
#3. Given a string of brackets, determine whether each sequence of brackets is balanced. If a string is balanced, return True otherwise, return False
def is_balanced(string):
    stack = []
    opening_brackets = {'{', '[', '('}
    closing_brackets = {']': '[', '}': '{', ')': '('}

    for char in string:
        if char in opening_brackets:
            stack.append(char)
        elif char in closing_brackets:
            if not stack or stack.pop() != closing_brackets[char]:
                return False
    return not stack
string1 = '{[(]}'
print(is_balanced(string1))
string2 = '{[]}'
print(is_balanced(string2))
```

```python

#4. Reverse the first ‘k’ elements of a queue, push it back into the queue.

from collections import deque

def reverse_k_elements(queue, k):
    if not queue or k <= 0 or k > len(queue):
        return queue

    stack = []
    for _ in range(k):
        stack.append(queue.popleft())

    while stack:
        queue.append(stack.pop())

    return queue

my_queue = deque([10, 20, 30, 40, 50, 60, 70, 80, 90])
k = 5
result_queue = reverse_k_elements(my_queue, k)
print(result_queue)
```
