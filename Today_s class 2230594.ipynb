```python
# Problem : Write a function that takes a
#number as input and prints if the number is a
#multiple of 2 or 5 or both or none.
def is_multiple(n):
#calculations
    if n % 2 == 0 and n % 5 == 0:
        print('Multiple of 2 and 5')
    elif n % 2 == 0:
        print('Multiple of 2')
    elif n % 5 ==0:
        print('Multiple of 5')
    else:
        print ('None')

#checking if the number is a multiple

is_multiple(10)
is_multiple(15)
is_multiple(22)
is_multiple(11)
```
