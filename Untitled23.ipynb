```python
lst = [2, -1, 4, 5, 2, -2, -3, 6, -2, 5]

def mystery(lst, i=0):
    if i < 0 or i >= len(lst):
        return

    next_index = i + lst[i]
    lst[i] = 100
    mystery(lst, next_index)

# driver code
mystery(lst)
print(lst)

```

```python
def rec(n):
    print(n, end=' ')
    if n == 1:
        return
    if n % 2 == 0:
        rec(n // 2)
    else:
        rec(n * 3 + 1)

rec(13)
```
