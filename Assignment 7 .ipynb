```python
#main codes



class graphAdjacencyList:
  def __init__(self,dir = False):
    self.dir = dir
    self.graph = dict()


  def addNode(self,node):
    self.graph[node] = []

  def addEdge(self,st,end):
    if self.graph.get(st) != None and self.graph.get(end) != None:
      if end not in self.graph[st]:
        self.graph[st].append(end)
        if not self.dir:
          if st not in self.graph[end]:
            self.graph[end].append(st)

  def removeEdge(self,st,end):
    if self.graph.get(st) != None and self.graph.get(end) != None:
      self.graph[st].remove(end)
      if not self.dir:
        self.graph[end].remove(st)

  def removeNode(self, node):
    if self.graph.get(node) != None:
      for key in self.graph:
        if node in self.graph[key]:
          self.graph[key].remove(node)
      del self.graph[node]

  def getOutDegree(self):
    outDegree = {}
    for i in self.graph:
      outDegree[i] = len(self.graph[i])
    return outDegree


  def findMaxInOutDegree(self):
    indegree = self.getInDegree()
    outdegree = self.getOutDegree()
    maxIn = -1
    maxOut = -1

    for i in indegree:
      if indegree[i] > maxIn:
        maxIn = indegree[i]
    print('Nodes with maximum in Degree: ', end = ' ')
    for i in indegree:
      if indegree[i] == maxIn:
        print(i, end= ' ')
    print()
    for j in outdegree:
      if outdegree[j] > maxOut:
        maxOut = outdegree[j]
    print('Nodes with maximum out Degree: ', end = ' ')
    for j in outdegree:
      if outdegree[j] == maxOut:
        print(j, end= ' ')
    print()

  def getInDegree(self):
    inDegree = {}
    merged_list = []
    for i in self.graph:
      merged_list += self.graph[i]

    for j in self.graph:
      counter = 0
      for k in merged_list:
        if j == k:
          counter += 1
      inDegree[j] = counter
    return inDegree

  def printAdjacencyList(self):
    for node in self.graph:
      print(f'{node}: {self.graph[node]}')


  def printSinkNodes(self):
    print('Sink nodes are: ', end = '-->')
    for i in self.getOutDegree():
      if self.getOutDegree()[i] == 0:
        print(i, end = ' ')


  def findCommonAdjacentNodes(graph, node1, node2):
      if not graph or node1 < 0 or node1 >= len(graph) or node2 < 0 or node2 >= len(graph):
          return "Invalid input"
      adjacent_nodes_node1 = []
      adjacent_nodes_node2 = []
      for i in range(len(graph[node1])):
          if graph[node1][i] == 1:
              adjacent_nodes_node1.append(i)
      for i in range(len(graph[node2])):
          if graph[node2][i] == 1:
              adjacent_nodes_node2.append(i)
      common_adjacent_nodes = set(adjacent_nodes_node1) & set(adjacent_nodes_node2)
      return list(common_adjacent_nodes)
```
