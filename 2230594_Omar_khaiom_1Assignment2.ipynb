```python
#Problem 8: Pizza Party with Friends ðŸ˜‡

slices = int(input("Enter the number of slices: "))
total_price = float(input("Enter the total price of the pizza: "))
your_slices = int(input("Enter the number of slices you ate: "))

friends_slices = []
for i in range(5):
    friend_slices = int(input(f"Enter the number of slices friend {i+1} ate: "))
    friends_slices.append(friend_slices)

total_slices_eaten = your_slices + sum(friends_slices)
price_per_slice = total_price / slices

print("\nAmount each should pay:")
print("You:", your_slices * price_per_slice)

friend_number = 1
for slices in friends_slices:
    friend_payment = slices * price_per_slice
    print("Friend", friend_number, ":", friend_payment)
    friend_number += 1
```

```python
#Problem 9: Distance in 2D Space
def calculate_distance(a, b, c, d):
    distance = ((c - a) ** 2 + (d - b) ** 2) ** 0.5
    return distance

a = float(input("Enter the x-coordinate of the first point: "))
b = float(input("Enter the y-coordinate of the first point: "))
c = float(input("Enter the x-coordinate of the second point: "))
d = float(input("Enter the y-coordinate of the second point: "))
distance = calculate_distance(a, b, c, d)
print("Distance:", "%.4f" % distance)
```

```python
#Problem 10: Complex Number Multipication
def multiply_complex(x, y, u, v):
    real_part = x * u - y * v
    imaginary_part = x * v + y * u
    return real_part, imaginary_part
x = float(input("Enter the real part of the first complex number: "))
y = float(input("Enter the imaginary part of the first complex number: "))
u = float(input("Enter the real part of the second complex number: "))
v = float(input("Enter the imaginary part of the second complex number: "))
result_real, result_imaginary = multiply_complex(x, y, u, v)
print("The multiplication of the two complex numbers is:", result_real, "+", result_imaginary, "i")
```
