```python
class stack1():
  def __init__(self):
    self.data = []
  def size(self):
    return len(self.data)
  def isempty(self):
    if self.size() == 0:
      return True
    else:
      return False
  def push(self,val):
      self.data.append(val)
  def pop(self):
    if self.isempty():
      return "NO data in the stack"
    else:
      self.data.pop()
  def peek(self):
    if self.isempty():
      return "NO data in the stack"
    else:
      return self.data[-1]
  def out(self):
    print(self.data)
```

```python
k=stack1()
a=input("")
for i in a:
  k.push(i)
while not k.isempty():
  val=k.peek()
  print(f"{val} : {ord(val)}")
  k.pop()
```

```python
class stack3():
  def __init__(self):
    self.data = []
  def size(self):
    return len(self.data)
  def isempty(self):
    if self.size() == 0:
      return True
    else:
      return False
  def push(self,val):
      self.data.append(val)
  def pop(self):
    if self.isempty():
      return "NO data in the stack"
    else:
      self.data.pop()
  def peek(self):
    if self.isempty():
      return "NO data in the stack"
    else:
      return self.data[-1]
  def out(self):
    print(self.data)
```

```python
s = input("Enter a string of brackets: ")
m = stack3()
for c in s:
    if c in "({[":
        m.push(c)
    else:
        if not m.data:
            print("unbalced")
        if c == ")" and m.peek() == "(":
            m.pop()
        elif c == "}" and m.peek() == "{":
            m.pop()
        elif c == "]" and m.peek() == "[":
            m.pop()
        else:
            break

if m.isempty():
  print("balanced")
else:
  print("unbalanced")
```

```python
class Stack4():
    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)
    
    def isempty(self):
        if self.size() == 0:
            return True
        else:
            return False
        
    def push(self, item):
        self.data.append(item)

    def pop(self):
        if self.isempty():
            print("Stack is empty")
        else:
            self.data.pop()
        
    def peek(self):
        if self.isempty():
            print("Stack is empty")
        else:
            return self.data[-1]
        
    def display(self):
        if self.isempty():
            print("Stack is empty")
        else:
            print(self.data[::-1])
    
    def keep_largest_on_top(self):
      h=Stack4()
      max=self.peek()
      while not self.isempty():
        val=self.peek()
        if val>max:
          max=val
        h.push(val)
        self.pop()

      while not h.isempty():
        val=h.peek()
        if val!=max:
          self.push(val)
        h.pop()
      self.push(max)
      self.display()

y=Stack4()
y.push(23)
y.push(53)
y.push(56)
y.push(97)
y.push(44)
y.push(22)
y.push(91)
y.push(1000)
y.keep_largest_on_top()
```

```python
class Queue():
    def __init__(self):
        self.data = []

    def isempty(self):
        return not len(self.data)

    def enqueue(self, element):
        self.data.append(element)

    def dequeue(self):
        if self.isempty():
            return "Warning: The queue is empty"
        self.data.pop(0)

    def peek(self):
        if self.isempty():
            return "Warning: The queue is empty"
        return self.data[0]
    
    def display(self):
       if self.isempty():
        print("Queue is empty")
       else:
        print(f"{self.data}")  
p=Queue()
A=Queue()
B=Queue()
k=[2 ,4 ,7 ,10 ,1 ,9 ,11]
for i in k:
  p.enqueue(i)

for m in p.data:
  if m%2==0:
    A.enqueue(m)
  else:
    B.enqueue(m)
print("A : ",end="")
A.display()
print("B : ",end="")
B.display() 
```
