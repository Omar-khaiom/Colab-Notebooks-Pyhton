```python
#main codes

import collections
class Node:
  def __init__(self, data):
    self.data = data
    self.right_child = None
    self.left_child = None
    self.numOfDescendants = 0

class BinarySearchTree:
  def __init__(self):
    self.root = None

  def isDuplicate(root1, root2):
      if (root1 == None and root2 == None) :
          return True

      elif (root1 != None and root2 == None) :
          return False
      elif (root1 == None and root2 != None) :
          return False
      else:
          return (root1.data == root2.data and
              isDuplicate(root1.left_child, root2.left_child)
              and isDuplicate(root1.right_child, root2.right_child))
  def insert(self, data):
    newnode = Node(data)
    if self.root is None:
      self.root = newnode
    else:
      current = self.root
      parent = None
      while current:
        parent = current
        if newnode.data <= current.data:
          current = current.left_child
        else:
          current = current.right_child

      if  newnode.data <= parent.data:
          parent.left_child = newnode
      else:
        parent.right_child = newnode

  def inorder(self, root_node):
    current = root_node
    if current is None:
      return
    else:
      self.inorder(current.left_child)
      print(current.data, end = ",")
      self.inorder(current.right_child)

  def countNumOfDesecendants(root):
      if root is None:
          return 0
      else:
          leftCount = countNumOfDesecendants(root.left_child)
          rightCount = countNumOfDesecendants(root.right_child)
          root.numOfDescendants = leftCount + rightCount
      return root.numOfDescendants + 1

  def isleafoddeven(self,root):
    current=root
    if current is None:
      return
    else:
      self.isleafoddeven(current.left_child)
      if (current.left_child is None and current.right_child is None):
        if (current.data %2 ==1):
          print(f"{current.data} is odd",end= ",")
        else:
          print(f"{current.data} is even ",end =",")
    self.isleafoddeven(current.right_child)
  def printlevelsinlevel(self, root):
    ans = []

    if root is None:
        return ans

    queue = collections.deque()
    queue.append(root)

    while queue:

        currSize = len(queue)
        currList = []

        while currSize > 0:

            currNode = queue.popleft()
            currList.append(currNode.data)
            currSize -= 1

            if currNode.left_child is not None:
                queue.append(currNode.left_child)

            if currNode.right_child is not None:
                queue.append(currNode.right_child)
        ans.append(currList)
    return ans
    def calculateSumR(self, root):
      if root is None:
        return 0
      else:
        return root.data + self.calculateSumR(root.left_child) + self.calculateSumR(root.right_child)

  def calculateSumR(self, root):
    if root is None:
      return 0
    else:
      return root.data + self.calculateSumR(root.left_child) + self.calculateSumR(root.right_child)

  def calculateSumIter(self, root):
    current = root
    stack = []
    count = 0
    while True:
        if current is not None:
            stack.append(current)
            current = current.left_child

        elif(stack):
            current = stack.pop()
            count = count + current.data
            current = current.right_child
        else:
            break
    return count


  def countNodesR(self, root):
    if root is None:
      return 0
    else:
      return 1+ self.countNodesR(root.left_child) + self.countNodesR(root.right_child)








  def countNodesIter(self, root):
    current = root
    stack = []
    count = 0
    while True:
        if current is not None:
            stack.append(current)
            current = current.left_child

        elif(stack):
            current = stack.pop()
            count = count + 1
            current = current.right_child
        else:
            break
    return count
```

```python
#6. Write a Function bool isBST(root) to check whether the tree with root is a
#valid BST or not. Please write the function iteratively and recursively.

def find_min(root):
    current = root
    while current.left_child:
      current = current.left_child
    return current

def find_max(root):
    current = root
    while current.right_child:
      current = current.right_child
    return current
def isBST(root):

     if (root == None):
        return True
     if (root.left_child != None and find_max(root.left_child).data > root.data ):
        return False
     if (root.right_child != None and find_min(root.right_child).data < root.data):
        return False
     return  (isBST(root.left_child) and isBST(root.right_child))
def isBSTiter(root):
    Stack = []
    prev = None

    while (Stack or root):
         while root:
            Stack.append(root)
            root = root.left_child
         root = Stack.pop()

         if (prev and root.data <= prev.data):
            return False

         prev = root
         root = root.right_child

    return True

bst = BinarySearchTree()
num = [7, 15, 9, 8, 10, 5, 3, 16, 11, 13,  17]
for i in range (0,len(num)):
     bst.insert(num[i])
print("Inorder with recursion BST:")
bst.inorder(bst.root)
print("")
print(isBST(bst.root))
print(isBSTiter(bst.root))
```
