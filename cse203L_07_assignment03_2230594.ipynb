```python
#Binary Search Tree – mainCodes

from collections import deque
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
class BSTree:
    def __init__(self):
        self.root = None
    def insert(self, value):
        self.root = self.insertR(self.root, value)
    def insertR(self, node, value):
        if not node:
            node = TreeNode(value)
            return node
        else:
            if value <= node.data:
                node.left = self.insertR(node.left, value)
            else:
                node.right = self.insertR(node.right, value)
            return node
    def inorder(self, node):
        if node is None:
            return
        else:
            self.inorder(node.left)
            print(node.data, end=' ')
            self.inorder(node.right)
    def print_leaf_nodes(self, node):
        if node is None:
            return
        else:
            self.print_leaf_nodes(node.left)
            if node.left is None and node.right is None:
                print(node.data)
            self.print_leaf_nodes(node.right)
    def find_node_height(self, seek_value):
        return self.find_node_heightR(self.root, seek_value)
    def find_node_heightR(self, node, seek_value, current_height=0):
        if node is None:
            return None
        if node.data == seek_value:
            return current_height
        elif seek_value < node.data:
            return self.find_node_heightR(node.left, seek_value, current_height + 1)
        else:
            return self.find_node_heightR(node.right, seek_value, current_height + 1)
    def find_deepest_node(self):
        res = [-1]
        maxLevel = [-1]
        self.find(self.root, 0, maxLevel, res)
        return res[0]
    def find(self, root, level, maxLevel, res):
        if root is not None:
            level += 1
            self.find(root.left, level, maxLevel, res)
            if level > maxLevel[0]:
                res[0] = root.data
                maxLevel[0] = level
            self.find(root.right, level, maxLevel, res)
    def max_width_bst(self):
        if self.root is None:
            return []
        max_width_nodes = []
        max_width = 0
        queue = deque([(self.root, 0)])
        while queue:
            level_size = len(queue)
            if level_size > max_width:
                max_width = level_size
                max_width_nodes = [node.data for node, _ in queue]
            for _ in range(level_size):
                current_node, position = queue.popleft()
                if current_node.left:
                    queue.append((current_node.left, position * 2))
                if current_node.right:
                    queue.append((current_node.right, position * 2 + 1))
        return max_width_nodes
    def find_depth(self, root, data, depth=0):
        if root is None:
            return -1
        if root.data == data:
            return depth
        elif data < root.data:
            return self.find_depth(root.left, data, depth + 1)
        else:
            return self.find_depth(root.right, data, depth + 1)
    def print_nodes_by_level(self):
        if not self.root:
            return
        result = []
        queue = deque([self.root])
        while queue:
            level_size = len(queue)
            current_level = []
            for _ in range(level_size):
                current_node = queue.popleft()
                current_level.append(current_node.data)
                if current_node.left:
                    queue.append(current_node.left)
                if current_node.right:
                    queue.append(current_node.right)
            result.append(current_level)
        for i, level in enumerate(result):
            print(f"Level {i + 1}: {level}")
    def delete_node(self, key):
        self.root = self._delete_node(self.root, key)
    def _delete_node(self, root, key):
        if root is None:
            return root
        if key < root.data:
            root.left = self._delete_node(root.left, key)
        elif key > root.data:
            root.right = self._delete_node(root.right, key)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            root.data = self._get_min_value_node(root.right).data
            root.right = self._delete_node(root.right, root.data)
        return root
    def _get_min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current
    def are_siblings(self, root, node1, node2):
        if root is None:
            return False

        if (root.left and root.right) and (
            (root.left.data == node1 and root.right.data == node2) or
            (root.right.data == node1 and root.left.data == node2)
        ):
            return True
        return (
            self.are_siblings(root.left, node1, node2) or
            self.are_siblings(root.right, node1, node2)
        )

    def kth_smallest_element(self, k):
        self.visited_count = 0
        self.kth_smallest = None
        self._inorder_traversal(self.root, k)
        return self.kth_smallest
    def _inorder_traversal(self, node, k):
        if node is None or self.kth_smallest is not None:
            return
        self._inorder_traversal(node.left, k)
        self.visited_count += 1
        if self.visited_count == k:
            self.kth_smallest = node.data
            return
        self._inorder_traversal(node.right, k)
def print_ancestors(root, target):
    if root is None:
        print("Node not found!")
        return False
    stack = []
    while root:
        stack.append(root)
        if target == root.data:
            break
        elif target < root.data:
            root = root.left
        else:
            root = root.right
    if root is None:
        print("Node not found!")
        return False
    print("Ancestors:", end=" ")
    for node in stack[:-1]:
        print(node.data, end=" ")
    print()
    return True
def print_successors(root, target):
    if root is None:
        print("Node not found!")
        return False
    found = False
    while root:
        if target == root.data:
            found = True
            break
        elif target < root.data:
            root = root.left
        else:
            root = root.right
    if not found:
        print("Node not found!")
        return False
    print("Successors:", end=" ")
    if root.right:
        current = root.right
        while current.left:
            current = current.left
        print(current.data, end=" ")
    stack = []
    while root:
        if target < root.data:
            stack.append(root.data)
            root = root.left
        elif target > root.data:
            root = root.right
        else:
            break
    if stack:
        print(", ".join(map(str, stack)))
    else:
        print("No successors")
    return True
```

```python
#main

from itertools import count

def adjacencylst_into_matrix(graph):
  matrix_elements=sorted(graph.keys())
  cols = rows =len(matrix_elements)
  adj_matrix=[[0 for x in range(rows)] for y in range(cols)]
  edges_list=[]

  for key in matrix_elements:
    for neighbors in graph[key]:
      edges_list.append((key,neighbors))

  for edge in edges_list:
      index_of_start_vertex=matrix_elements.index(edge[0])
      index_of_end_vertex=matrix_elements.index(edge[1])
      adj_matrix[index_of_start_vertex][index_of_end_vertex] =1

  for x in adj_matrix:
    print(x)

def adjacencyMatrix_into_list(adj_matrix):
    graph ={}

    for i in range(len(adj_matrix)):
      vertex =(str(i))
      neighbors=[]


      for j in range(len(adj_matrix[i])):
          if adj_matrix[i][j]==1:
              neighbors.append(str(j))

      graph[vertex]=neighbors

    return graph


def path_exists(graph, path, index=0):
    if index == len(path) - 1:
        return True

    current_vertex = path[index]
    next_vertex = path[index + 1]
    if next_vertex not in graph.get(current_vertex, []):
        return False
    return path_exists(graph, path, index + 1)


def nof_bidirectional_edges(graph):
    count = 0

    for vertex in graph:
      for neighbors in graph[vertex]:
          if vertex in graph.get(neighbors,[]):
            count+=1
    return count //2

def directed_convert_undirected(directed_graph):
    undirected_graph ={}

    for vertex in directed_graph:
        if vertex not in undirected_graph:
          undirected_graph[vertex]=[]

        for neighbors in directed_graph[vertex]:
            undirected_graph[vertex].append(vertex)

            if neighbors not in undirected_graph:
              undirected_graph[neighbors]=[]
            undirected_graph[neighbors].append(vertex)
    return undirected_graph


def remove_bidirectional_edges(graph):
    has_bidirectional_edges = False
    for vertex in graph:
        for neighbor in graph[vertex]:
            if vertex in graph.get(neighbor, []):
                if neighbor in graph[vertex]:
                    graph[vertex].remove(neighbor)
                if vertex in graph[neighbor]:
                    graph[neighbor].remove(vertex)
                has_bidirectional_edges = True
    if not has_bidirectional_edges:
        print("No bidirectional edge found")
def remove_even_nodes(graph):
    odd_graph = {}

    for vertex in graph:
        if vertex % 2 != 0:  # Check if the vertex is odd
            odd_neighbors = [neighbor for neighbor in graph[vertex] if neighbor % 2 != 0]
            odd_graph[vertex] = odd_neighbors

    return odd_graph

def count_edges(graph):
    edge_count = 0
    for vertex in graph:
        edge_count += len(graph[vertex])
    first_vertex = next(iter(graph), None)
    if first_vertex is not None and type(graph[first_vertex]) == list:
        edge_count //= 2
    return edge_count

def common_adjaNodes(graph, node1, node2, node3):
    neighbors_node1 = set(graph.get(node1, []))
    neighbors_node2 = set(graph.get(node2, []))
    neighbors_node3 = set(graph.get(node3, []))
    common_neighbors = neighbors_node1.intersection(neighbors_node2, neighbors_node3)
    return list(common_neighbors)

def source_and_sink(matrix):
    num_nodes = len(matrix)
    sources = []
    sinks = []
    for i in range(num_nodes):
        if all(matrix[j][i] == 0 for j in range(num_nodes) if j != i):
            sources.append(i)
        if all(matrix[i][j] == 0 for j in range(num_nodes) if j != i):
            sinks.append(i)
    print("Source Nodes:", sources)
    print("Sink Nodes:", sinks)
```
