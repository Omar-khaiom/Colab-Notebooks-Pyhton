Stack Implementation

```python
class Stack:
    def __init__(self):
        self.__data = []

    def size(self):
        return len(self.__data)

    def isempty(self):
      if self.size()==0:
        return True
      else:
        return False

    # Push data to the top of the stack
    def push(self, data):
        self.__data.append(data)

    def pop(self):
      if self.isempty():
        print("Stack is empty")
      else:
         self.__data.pop()



    def peek(self):
      if self.isempty():
        print("Stack is empty")
      else:
        return self.__data[-1]
    def clearTahmid(self):
       if self.isempty():
        print("Stack is empty")
       else:
        self.__data.clear()

#Optional/Extra function. Just only for display. No use for stack functionality
    def printStack(self):
       if self.isempty():
        print("Stack is empty")
       else:
        print(self.__data)
```

```python
s = Stack()
#print(s.isempty())
s.peek()
s.push('B')
s.push('A')
s.push('C')
s.push('D')
s.printStack()
s.pop()
s.printStack()
s.push(10)
s.printStack()
s.push(30)
s.push(50)
s.printStack()
print(s.peek())
s.pop()
s.pop()
print(s.peek())
s.push(10)
s.push(30)

s.printStack()
```

Practice Problem 3: Suppose, you have implemented your stack class. You have 5 functions: isempty(), push(), pop(), peek() and size(). Write a function named ‘remove_duplicates(input_string)’ using this stack class to remove all the adjacent (পাশাপাশি) duplicate characters from a given string. (4 marks) Example: remove_duplicates(“aabcbtt”) will return bcb.

```python
class Stack:

    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isempty(self):
      if self.size()==0:
        return True
      else:
        return False

    # Push data to the top of the stack
    def push(self, data):
        self.data.append(data)


    def pop(self):
      if self.isempty():
        print("Stack is empty")
      else:
         self.data.pop()



    def peek(self):
      if self.isempty():
        print("Stack is empty")
      else:
        return self.data[-1]




#Optional/Extra function. Just only for display. No use for stack functionality
    def printStack(self):
       if self.isempty():
        print("Stack is empty")
       else:
        print(self.data)

def remove_duplicates(input_string):
  res = Stack()
  for ch in input_string:
    if not res.isempty() and res.peek() == ch:
      res.pop()
    else:
      res.push(ch)

  return res

s = remove_duplicates("aabcbtt")
s.printStack()
resultStr = print("".join(s.data))
```

Reverse String:

```python
def reverseString(str):
  s = Stack()
  for ch in str:
    s.push(ch)

  rev = ""
  for i in range (0, s.size()):
    ch = s.peek()
    rev += ch
    s.pop()
  return rev

print(reverseString("WATERMELON"))
```

Implementing Circular Queue:

1. In case of a circular queue, head pointer will always point to the front of the queue, and tail pointer will always point to the end of the queue.
2. Initially, the head and the tail pointers will be pointing to the same location, this would mean that the queue is empty.

circular-queue-1.png

3. New data is always added to the location pointed by the tail pointer, and once the data is added, tail pointer is incremented to point to the next available location.

circular-queue-2.png

4. In a circular queue, data is not actually removed from the queue. Only the head pointer is incremented by one position when dequeue is executed. As the queue data is only the data between head and tail, hence the data left outside is not a part of the queue anymore, hence removed.

circular-queue-3.png

5. The head and the tail pointer will get reinitialised to 0 every time they reach the end of the queue.
circular-queue-4.png

6. Also, the head and the tail pointers can cross each other. In other words, head pointer can be greater than the tail. Sounds odd? This will happen when we dequeue the queue a couple of times and the tail pointer gets reinitialised upon reaching the end of the queue.

circular-queue-5.png

7. Another very important point is keeping the value of the tail and the head pointer within the maximum queue size. In the diagrams above the queue has a size of 8, hence, the value of tail and head pointers will always be between 0 and 7. This can be controlled either by checking everytime whether tail or head have reached the maxSize and then setting the value 0 or, we have a better way, which is, for a value x if we divide it by 8, the remained will never be greater than 8, it will always be between 0 and 7, which is exactly what we want.

```python
#Optional Topic
#This is the CircularQueue class
class CircularQueue:

  # constructor for the class
  # taking input for the size of the Circular queue
  # from user
  def __init__(self, maxSize):
    self.queue = list()
    # user input value for maxSize
    self.maxSize = maxSize
    self.head = 0
    self.tail = 0

  # add element to the queue
  def enqueue(self, data):
    # if queue is full
    if self.size() == (self.maxSize - 1):
      return("Queue is full!")
    else:
      # add element to the queue
      self.queue.append(data)
      # increment the tail pointer
      self.tail = (self.tail+1) % self.maxSize
      return True

  # remove element from the queue
  def dequeue(self):
    # if queue is empty
    if self.size() == 0:
      return("Queue is empty!")
    else:
      # fetch data
      data = self.queue[self.head]
      # increment head
      self.head = (self.head+1) % self.maxSize
      return data

        # peek element from the queue
  def peek(self):
    # if queue is empty
    if self.size() == 0:
      return("Queue is empty!")
    else:
      # fetch data
      data = self.queue[self.head]
      return data

  # find the size of the queue
  def size(self):
    if self.tail >= self.head:
      qSize = self.tail - self.head
    else:
      qSize = self.maxSize - (self.head - self.tail)
    # return the size of the queue
    return qSize

# input 7 for the size or anything else
size = input("Enter the size of the Circular Queue")
q = CircularQueue(int(size))

# change the enqueue and dequeue statements as you want
print(q.enqueue(10))
print(q.enqueue(20))
print(q.enqueue(30))
print(q.enqueue(40))
print(q.enqueue(50))
print(q.enqueue('Studytonight'))
print(q.enqueue(70))
print(q.enqueue(80))
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
```

Full Standard Queue Implementation:

```python
from collections import deque

# Queue implementation using deque (Double ended queue)
class Queue:
    def __init__(self):
        self.__data = deque()

    def size(self):
        return len(self.__data)

    def isempty(self):
      if self.size()==0:
        return True
      else:
        False

    def enqueue(self, data):
        self.__data.append(data)

    def dequeue(self):
      if self.isempty():
        print("Queue is empty")
      else:
        self.__data.popleft()

    def peek(self):
      if self.isempty():
        print("Queue is empty")
      else:
        return self.__data[0]

    def printQueue(self):
       if self.isempty():
        print("Queue is empty")
       else:
        print(self.__data)
```

```python
q = Queue()
q.peek()
q.dequeue()
q.enqueue("Aquib")
q.enqueue("Arif")
q.enqueue("Sajib")
q.enqueue("Nafis")
#q.peek()
q.dequeue()
q.printQueue()
```

Media player using queue

```python
from random import randint

class Track:
  minLength = 5
  maxLength = 10

  def __init__(self, title):
    self.title = title
    self.length = randint(Track.minLength, Track.maxLength)
```

```python
t1 = Track("Lemon Tree")
t2 = Track("Happy")
print("Title: {} Length: {}".format(t1.title, t1.length))
print("Title: {} Length: {}".format(t2.title, t2.length))
```

```python
import time

q = Queue()
q.enqueue(Track("Song A"))
q.enqueue(Track("Song B"))
q.enqueue(Track("Song C"))


def play():
  while not q.isempty():
    current_song = q.peek()
    print("Playing {}...",current_song.title)
    time.sleep(current_song.length)
    q.dequeue()
  print("All tracks finished.")


play()
```
