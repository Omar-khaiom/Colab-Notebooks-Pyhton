```python
#1
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def getAverage(head):
    total_sum = 0
    count = 0
    current = head
    while current:
        total_sum += current.data
        count += 1
        current = current.next
    return total_sum / count

head=Node(5)
head.next=Node(6)
head.next.next=Node(7)
head.next.next.next=Node(8)
print(getAverage(head))
```

```python
#2
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def numOfOccurrences(self, value):
    count = 0
    current = self
    while current:
        if current.data == value:
            count += 1
        current = current.next
    return count

head=Node(2)
head.next=Node(5)
head.next.next=Node(2)
head.next.next.next=Node(8)
head.next.next.next.next=Node(2)
print(numOfOccurrences(head, 2))
```

```python
#3
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def compare_lists(head1, head2):

    while head1 and head2:
        if head1.data != head2.data:
            return False
        head1 = head1.next
        head2 = head2.next


    if not head1 and not head2:
        return True
    else:
        return False


list1=Node(7)
list1.next=Node(6)
list1.next.next=Node(10)
list1.next.next.next=Node(11)

list2=Node(7)
list2.next=Node(6)
list2.next.next=Node(10)
list2.next.next.next=Node(11)

list3=Node(7)
list3.next=Node(16)
list3.next.next=Node(10)
list3.next.next.next=Node(11)

list4=Node(7)
list4.next=Node(6)
list4.next.next=Node(10)
list4.next.next.next=Node(11)


print(compare_lists(list1, list2))
print(compare_lists(list1, list3))
print(compare_lists(list1, list4))
```

```python
#4
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def addNodeBeforeValue(self, givenValue, newValue):

        if self.head is None:
            print("Not found")
            return


        if self.head.data == givenValue:
            newNode = Node(newValue)
            newNode.next = self.head
            self.head = newNode
            return

        current = self.head
        while current.next is not None:
            if current.next.data == givenValue:
                newNode = Node(newValue)
                newNode.next = current.next
                current.next = newNode
                return
            current = current.next


        print("Not found")

#test_the_function
sll = SinglyLinkedList()
sll.head = Node(3)
sll.head.next = Node(6)
sll.head.next.next = Node(7)
sll.head.next.next.next = Node(10)

sll.addNodeBeforeValue(8, 6)
current = sll.head
while current is not None:
    print(current.data, end=" ")
    current = current.next


sll.addNodeBeforeValue(4, 1)
```

```python
#6

class Node:
	def __init__(self, data):
		self.data = data
		self.next = None

class LinkedList:


	def __init__(self):
		self.head = None

	def sortedInsert(self, new_node):


		if self.head is None:
			new_node.next = self.head
			self.head = new_node

		elif self.head.data >= new_node.data:
			new_node.next = self.head
			self.head = new_node

		else :

			current = self.head
			while(current.next is not None and
				current.next.data < new_node.data):
				current = current.next

			new_node.next = current.next
			current.next = new_node


	def push(self, new_data):
		new_node = Node(new_data)
		new_node.next = self.head
		self.head = new_node

	def printList(self):
		temp = self.head
		while(temp):
			print(temp.data, end=" ")
			temp = temp.next


# Driver program
llist = LinkedList()
new_node = Node(3)
llist.sortedInsert(new_node)
new_node = Node(5)
llist.sortedInsert(new_node)
new_node = Node(6)
new_node = Node(7)
llist.sortedInsert(new_node)
new_node = Node(10)
llist.sortedInsert(new_node)
print("Create Linked List")
llist.printList()

```

```python
#7


class Node:
	def __init__(self, data):
		self.data = data
		self.next = None


class LinkedList:


	def __init__(self):
		self.head = None


	def reverse(self):
		prev = None
		current = self.head
		while(current is not None):
			next = current.next
			current.next = prev
			prev = current
			current = next
		self.head = prev


	def push(self, new_data):
		new_node = Node(new_data)
		new_node.next = self.head
		self.head = new_node


	def printList(self):
		temp = self.head
		while(temp):
			print(temp.data, end=" ")
			temp = temp.next



llist = LinkedList()
llist.push(3)
llist.push(10)
llist.push(17)
llist.push(61)
llist.push(117)
print ("Given list")
llist.printList()
llist.reverse()
print ("\nReversed list")
llist.printList()

```
