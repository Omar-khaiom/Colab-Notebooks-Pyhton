```python
#1. Write insert(self, data) function that will create a node and insert the node into a BST. You should write the insert function with iteration.
#Then Take a list of 12 elements. Randomly generate 12 numbers and put them into the list.
#Then in a loop, use insert function to insert the elements in to a BST


import random

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):
        new_node = Node(data)
        if not self.root:
            self.root = new_node
            return

        current = self.root
        while True:
            if data < current.val:
                if current.left is None:
                    current.left = new_node
                    break
                current = current.left
            else:
                if current.right is None:
                    current.right = new_node
                    break
                current = current.right

def main():
    bst = BST()
    random_numbers = random.sample(range(1, 100), 12)

    for num in random_numbers:
        bst.insert(num)
    print("Inorder Traversal of BST:")
    inorder(bst.root)

def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
if __name__ == "__main__":
    main()
```

```python
#6. Check whether the leaf nodes of a BST is even or odd.

#10

#/ \

#5 13

#/ \ \

#3 7 14

#The output of the BST above is [odd, odd, even].

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
def isLeaf(node):
    return node is not None and node.left is None and node.right is None
def checkLeafNodes(root):
    if not root:
        return []
    result = []
    def dfs(node):
        if isLeaf(node):
            result.append("even" if node.val % 2 == 0 else "odd")
            return
        if node.left:
            dfs(node.left)
        if node.right:
            dfs(node.right)
    dfs(root)
    return result
def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)

    leaf_nodes_parity = checkLeafNodes(root)
    print("The leaf nodes are:", leaf_nodes_parity)

if __name__ == "__main__":
    main()

```

```python
#7. Write a function int calculatesum(root) to calculate sum of all nodes of the BST with root.
#Please write the function iteratively and recursivel



class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def calculateSumIterative(root):
    if not root:
        return 0
    stack = [root]
    total_sum = 0
    while stack:
        current = stack.pop()
        total_sum += current.val
        if current.right:
            stack.append(current.right)
        if current.left:
            stack.append(current.left)
    return total_sum

def calculateSumRecursive(root):
    if not root:
        return 0

    return root.val + calculateSumRecursive(root.left) + calculateSumRecursive(root.right)

def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)
    sum_iterative = calculateSumIterative(root)
    print("Sum of all nodes (iterative):", sum_iterative)
    sum_recursive = calculateSumRecursive(root)
    print("Sum of all nodes (recursive):", sum_recursive)

if __name__ == "__main__":
    main()
```

```python
#8. Write a function int countNodes(root) to calculate number of all nodes

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
def countNodes(root):
    if not root:
        return 0
    return 1 + countNodes(root.left) + countNodes(root.right)
def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)
    num_nodes = countNodes(root)
    print("Number of nodes in the BST:", num_nodes)
if __name__ == "__main__":
    main()
```

```python
#9. Write a function int calculatesumleaf(root) to calculate sum of all leaf nodes of the BST with root.
# Please write the function iteratively and recursively.
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
def calculateSumLeafIterative(root):
    if not root:
        return 0
    stack = [root]
    leaf_sum = 0
    while stack:
        current = stack.pop()
        if not current.left and not current.right:
            leaf_sum += current.val
        if current.right:
            stack.append(current.right)
        if current.left:
            stack.append(current.left)
    return leaf_sum
def calculateSumLeafRecursive(root):
    if not root:
        return 0
    if not root.left and not root.right:
        return root.val
    return calculateSumLeafRecursive(root.left) + calculateSumLeafRecursive(root.right)
def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)
    sum_iterative = calculateSumLeafIterative(root)
    print("Sum of all leaf nodes (iterative):", sum_iterative)

    sum_recursive = calculateSumLeafRecursive(root)
    print("Sum of all leaf nodes (recursive):", sum_recursive)

if __name__ == "__main__":
    main()
```

```python
#10. Write a function int countLeafNodes(root) to calculate number of all leaf nodes
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def countLeafNodes(root):
    if not root:
        return 0
    if not root.left and not root.right:
        return 1
    return countLeafNodes(root.left) + countLeafNodes(root.right)

def main():
    root = Node(10)
    root.left = Node(5)
    root.right = Node(13)
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.right = Node(14)
    num_leaf_nodes = countLeafNodes(root)
    print("Number of leaf nodes in the BST:", num_leaf_nodes)
if __name__ == "__main__":
    main()
```
